# ● Compute player statistics and rankings based on these metrics:

# >>>batsmen's statistics and rankings based on different metrics for season year 2017: 

```python
print("batsmen's statistics and rankings based on avg runs:")
av=ba.groupby('Player').agg({'Avg': 'mean'}).sort_values(by='Avg',ascending=False).head(10)
av
```

```python
f, ax = plt.subplots(1,figsize=(10,6))
av['Avg'].plot.bar(color=['b', 'r', 'g', 'y', 'k'],ax=ax)
ax.set_ylabel('average runs')
ax.set_title('Average runs of Top 10 batsmen in 2017', y=1.02)
ax.set_xticklabels(av.index, rotation = 45);
```

```python
print("batsmen's statistics and rankings based on strike rate:")
sr=ba.groupby('Player').agg({'SR': 'mean'}).sort_values(by='SR',ascending=False).head(10)
sr
```

```python
f, ax = plt.subplots(1,figsize=(10,6))
sr['SR'].plot.bar(color=['b', 'r', 'g', 'y', 'k'],ax=ax)
ax.set_ylabel('Strike rate')
ax.set_title('Strike rate of Top 10 batsmen in 2017', y=1.02)
ax.set_xticklabels(sr.index, rotation = 45);
```

# 6. Validation and Sensitivity Analysis:

● Validate the player selection recommendations using statistical methods and by
comparing with successful IPL teams from previous seasons.

# ● Validate the player selection recommendations using statistical methods and by comparing with successful IPL teams from previous seasons:

```python
def validate_rating(rating):
    """"
    Description:
        validate the players if their POS is a digit between 1 to 5.

    Args:
        rating (str): rating to be validated

    Returns:
        True if rating is valid
        False if rating is not a valid
    """

    try:
        rating = int(rating)
        if rating < 1 or rating > 5:
            return False
        return True
    except ValueError:
        return False
```

```python
finaltm['POS'].apply(validate_rating)
```
